```python
music.drop(columns=['Rank', 'Peak' ,'All Time Peak', 'Ref.'],inplace=True)
```

```python
col_rename={'Actual\xa0gross':'gross', 'Adjusted\xa0gross (in 2022 dollars)':'adj_gross', 'Artist':'artist',
       'Tour title':'tour', 'Year(s)':'year', 'Shows':'shows', 'Average gross':'avg_gross'}
```

```python
music.rename(columns=col_rename,inplace=True)
```

```python
music.columns
```

```python
#how to decode
music.columns=music.columns.str.encode('ascii','ignore').str.decode('ascii')
```

```python
music.gross.sample(5)
```

```python
music.gross.str.replace('$','').str.replace(',','')
```

```python
cleancol=['gross','adj_gross','avg_gross']
```

```python
for col in cleancol:
  music[col]=music[col].str.replace(r'[^\d.]','',regex=True)
```

```python
music.gross
```

```python
music
```

```python
for col in cleancol:
      music[col]=pd.to_numeric(music[col])
```

```python
def extract_year(year):
  return year[-4:] if len(year)>4 else year
```

```python
music['year']=music.year.apply(extract_year)
```
