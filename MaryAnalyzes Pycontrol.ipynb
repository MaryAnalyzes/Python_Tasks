```python
# CREATE FUNCTION
def addition(w,y):
  return w+y
def subtract(w,y):
  return w-y
def multiply(w,y):
  return w*y
def Division(w,y):
  if y != 0:
     return w/y
  else:
    return "cannot divide by zero"
```

```python
addition (20,18)
```

```python
subtract(20,12)
```

```python
multiply(20,12)
```

```python
Division(20,0)
```

```python
#create dictionary named operations
operations ={'+':addition,'-':subtract,'*':multiply,'/':Division}
```

```python
# create function
def calculator():
  # prompt user to enter first No
    num1=float(input('First Number'))
    return num1
```

```python
num1=calculator()
print (num1)
```

```python
 while True:
       user_input=input("select'*','+','-','/' or 'quit' to exit:")
#promp user to select an operation symbol
       if user_input.lower()=='quit':
        break
       elif user_input in ['*','+','-','/']:
          num2=float(input('Enter Second Number'))
       else:
          print('unrecognized arithmetic expression')
          continue

       if user_input in operations:
        cal_function= operations[user_input]
        answer= cal_function(num1, num2)
        print(f"{num1} {user_input} {num2} ={answer}")
#print the equation and the result of cal
        num1=answer
 # update num1 with the result for further cal
       else:
        print(f" '{user_input}' is not allowed")

       should_continue= input("continue using the result as the first number for further calculation? (YES/NO/QUIT)")
       if should_continue.upper()== "NO":
          num1= float(input("Enter the first Number"))
          # start a new calculation
       elif should_continue.upper()=="QUIT":
        break
#otherwise continue using the result of the first number

```
